```python
import pandas as pd
import numpy as np
import datetime as dt
import panel as pn
pn.extension('tabulator')
pn.extension(sizing_mode="stretch_width")

import hvplot.pandas
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/bradleyhrc/glucose-data/main/BHC-Glucose-Data-Clean.csv')
```

```python
df['Device Timestamp'] = pd.to_datetime(df['Device Timestamp'], format='%d-%m-%Y %H:%M')
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')
df
# df['date_col'] =  pd.to_datetime(df['date_col'], format='%d/%m/%Y')
#pd.to_datetime(pd.Series(['05/23/2005']), format="%m/%d/%Y")
```

```python
# data already pre-processed, will add data preprocessing automation here in future update
# make data interactive
idf = df.interactive()
```

```python
date_picker = pn.widgets.DatePicker(name='Date Picker', start=dt.datetime(2019, 8, 12), end=dt.datetime(2021, 6, 5))
#pn.widgets.DatePicker(name='Date Picker', start=dt.datetime(2019, 8, 12), end=dt.datetime(2021, 6, 5), value=dt.datetime(2021, 6, 5))
date_picker
```

```python
yaxis_glucose = pn.widgets.RadioButtonGroup(
    name='Y axis',
    options=['Glucose mmol/L'],
    button_type='success'
)
yaxis_glucose
```

```python
glucose_pipeline = (
    idf[idf.Date == date_picker]
    .sort_values(by='Device Timestamp')
)

#glucose_pipeline = (
#    idf[(idf.Date <= str(date_slider))] # not done, left off
#    .to_frame()
#    .reset_index()
#    .sort_values(by='Date')
#)
```
